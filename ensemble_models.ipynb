```python
import os
import json
import hashlib
import joblib
import numpy as np
import pandas as pd
import gdown

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    classification_report, roc_auc_score, precision_recall_curve, auc, f1_score
)

from imblearn.over_sampling import SMOTE
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
```

## Machine Learning Pipeline

```python
"""
ForenXAI — Dev Side Pipeline (Acquisition -> Candidate Models)
=================================================================
Stages: acquisition, profiling, stratified subsample, preprocessing
(clean, encode), train/test split, SMOTE (train only), candidate
model training (DT, RF, Extra Trees, XGBoost, LightGBM) with a
Logistic Regression reference/meta-learner baseline.

Candidates are intentionally kept broad here and narrowed down
after CV comparison — see Section 7.
"""

# Cell 1 — config + acquisition function
RANDOM_STATE = 42
OUTPUT_DIR = "models"
os.makedirs(OUTPUT_DIR, exist_ok=True)

DRIVE_FILE_ID = "1dz9X40Xuz5--c2n7FCR3rSlLclDU763c"
RAW_CSV_PATH = "uwf_zeekdata24_combined.csv"
LABEL_COLUMN = "label_binary"

def acquire_dataset():
    if not os.path.exists(RAW_CSV_PATH):
        url = f"https://drive.google.com/uc?id={DRIVE_FILE_ID}"
        gdown.download(url, RAW_CSV_PATH, quiet=False)
    print(f"[+] Dataset acquired: {RAW_CSV_PATH}")
```

```python
# Cell 2 — the dtype-inspection function (just a definition, safe to run anytime)
def load_dataset():
    df = pd.read_csv(RAW_CSV_PATH, low_memory=False)
    known_object_cols = {"source_category", LABEL_COLUMN}
    suspect_cols = [c for c in df.select_dtypes(include="object").columns if c not in known_object_cols]
    for col in suspect_cols:
        print(f"[!] Column '{col}' is object-typed and not a known categorical column — inspect it")
        print(df[col].apply(type).value_counts())
    return df
```

```python
# Cell 3 — actually run them, in order
acquire_dataset()      # downloads the file first
df = load_dataset()    # then reads and inspects it
```

```python

# ============================================================
# 2. PROFILING
# Documents dataset shape before any modification — this becomes
# dataset_profile.json, a citable methodology artifact.
# ============================================================
def profile_dataset(df):
    profile = {
        "n_rows": len(df),
        "n_columns": df.shape[1],
        "columns": list(df.columns),
        "dtypes": {c: str(t) for c, t in df.dtypes.items()},
        "missing_values": df.isna().sum().to_dict(),
        "duplicate_rows": int(df.duplicated().sum()),
    }
    if LABEL_COLUMN in df.columns:
        profile["class_distribution"] = df[LABEL_COLUMN].value_counts().to_dict()

    with open(os.path.join(OUTPUT_DIR, "dataset_profile.json"), "w") as f:
        json.dump(profile, f, indent=2, default=str)

    print(f"[+] Profile: {profile['n_rows']:,} rows, {profile['n_columns']} columns")
    if "class_distribution" in profile:
        print(f"[+] Class distribution: {profile['class_distribution']}")
    return profile


# ============================================================
# 3. STRATIFIED SUBSAMPLE
# Reduces compute cost while preserving class proportions.
# ============================================================
def stratified_subsample(df, target_size=500_000, label_col=LABEL_COLUMN):
    if len(df) <= target_size:
        print(f"[+] Dataset already <= {target_size:,} rows, no subsampling needed")
        return df

    sampled, _ = train_test_split(
        df, train_size=target_size, stratify=df[label_col], random_state=RANDOM_STATE
    )
    print(f"[+] Subsampled {len(df):,} -> {len(sampled):,} rows (stratified)")
    return sampled


# ============================================================
# 4. PREPROCESSING — CLEANING
# Drops identifiers/leakage columns, fixes known mixed-type
# columns (history, service), removes inf/duplicates.
# ============================================================
def clean_data(df):
    before = len(df)

    df = df.drop(columns=[c for c in DROP_COLS if c in df.columns])

    if "history" in df.columns:
        df["history"] = df["history"].fillna("none").astype(str)
    if "service" in df.columns:
        df["service"] = df["service"].fillna("none").astype(str)

    df = df.replace([np.inf, -np.inf], np.nan)
    df = df.drop_duplicates()

    print(f"[+] Cleaning: {before:,} -> {len(df):,} rows after dedup")
    print(f"[+] Columns after drop: {df.shape[1]}")
    return df


def split_features_labels(df, label_col=LABEL_COLUMN):
    y = df[label_col]

    def coerce(val):
        if pd.isna(val):
            return "UNKNOWN"
        if isinstance(val, (bool, np.bool_)):
            return "Attack" if val else "Benign"
        if isinstance(val, (int, float, np.integer, np.floating)):
            return "Attack" if val == 1 else "Benign"
        return str(val).strip()

    y = y.apply(coerce)
    print("[+] Normalized label values:", y.value_counts().to_dict())

    X = df.drop(columns=[label_col])
    return X, y


# ============================================================
# 5. TRAIN / TEST SPLIT
# Stratified. Test set untouched until final evaluation.
# ============================================================
def split_data(X, y):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, stratify=y, random_state=RANDOM_STATE
    )
    print(f"[+] Split: {len(X_train):,} train / {len(X_test):,} test")
    return X_train, X_test, y_train, y_test


# ============================================================
# 6. ENCODING + SCALING (+ IMPUTATION)
# Fit on TRAIN ONLY. Imputation lives inside the pipeline so it
# is saved and reused consistently — never recomputed later.
# ============================================================
def build_and_fit_preprocessor(X_train):
    categorical_cols = X_train.select_dtypes(include=["object"]).columns.tolist()
    numeric_cols = X_train.select_dtypes(include=[np.number]).columns.tolist()

    numeric_pipeline = Pipeline([
        ("impute", SimpleImputer(strategy="median")),
        ("scale", StandardScaler()),
    ])
    categorical_pipeline = Pipeline([
        ("impute", SimpleImputer(strategy="most_frequent")),
        ("encode", OneHotEncoder(handle_unknown="ignore")),
    ])

    preprocessor = ColumnTransformer(transformers=[
        ("cat", categorical_pipeline, categorical_cols),
        ("num", numeric_pipeline, numeric_cols),
    ])
    preprocessor.fit(X_train)

    print(f"[+] Preprocessor fit: {len(categorical_cols)} categorical, {len(numeric_cols)} numeric columns")
    return preprocessor


# ============================================================
# 7. SMOTE — TRAIN SPLIT ONLY
# ============================================================
def apply_smote(X_train_enc, y_train):
    before_counts = pd.Series(y_train).value_counts().to_dict()
    sm = SMOTE(random_state=RANDOM_STATE)
    X_res, y_res = sm.fit_resample(X_train_enc, y_train)
    after_counts = pd.Series(y_res).value_counts().to_dict()
    print(f"[+] SMOTE: {before_counts} -> {after_counts}")
    return X_res, y_res


# ============================================================
# 8. CANDIDATE MODELS + CROSS-VALIDATION
# Broad pool now, narrowed after comparing CV scores (Section 9).
# ============================================================
def get_candidates():
    return {
        "decision_tree": DecisionTreeClassifier(random_state=RANDOM_STATE),
        "random_forest": RandomForestClassifier(n_estimators=300, random_state=RANDOM_STATE, n_jobs=-1),
        "extra_trees": ExtraTreesClassifier(n_estimators=300, random_state=RANDOM_STATE, n_jobs=-1),
        "xgboost": XGBClassifier(eval_metric="logloss", random_state=RANDOM_STATE, n_jobs=-1),
        "lightgbm": LGBMClassifier(random_state=RANDOM_STATE, n_jobs=-1),
        "logistic_regression": LogisticRegression(max_iter=2000, random_state=RANDOM_STATE),
    }


def train_and_cv_candidates(X_train_res, y_train_res, n_splits=5):
    candidates = get_candidates()
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=RANDOM_STATE)
    cv_results = {}

    for name, model in candidates.items():
        scores = cross_val_score(model, X_train_res, y_train_res, cv=skf, scoring="f1_macro", n_jobs=-1)
        cv_results[name] = {"mean_f1_macro": float(scores.mean()), "std": float(scores.std())}
        model.fit(X_train_res, y_train_res)
        print(f"[CV] {name}: F1-macro = {scores.mean():.4f} (+/- {scores.std():.4f})")

    with open(os.path.join(OUTPUT_DIR, "candidate_cv_results.json"), "w") as f:
        json.dump(cv_results, f, indent=2)

    return candidates, cv_results


# ============================================================
# 9. NARROW DOWN — TOP-N CANDIDATES BY CV SCORE
# ============================================================
def select_top_candidates(candidates, cv_results, top_n=3, exclude=("logistic_regression",)):
    eligible = {k: v for k, v in cv_results.items() if k not in exclude}
    ranked = sorted(eligible.items(), key=lambda x: x[1]["mean_f1_macro"], reverse=True)
    top_names = [name for name, _ in ranked[:top_n]]
    print(f"[+] Top {top_n} candidates: {top_names}")
    return {name: candidates[name] for name in top_names}


# ============================================================
# 10. SOFT-VOTING ENSEMBLE FROM TOP CANDIDATES
# ============================================================
def build_ensemble(top_candidates, X_train_res, y_train_res):
    ensemble = VotingClassifier(
        estimators=[(name, model) for name, model in top_candidates.items()],
        voting="soft"
    )
    ensemble.fit(X_train_res, y_train_res)
    return ensemble


# ============================================================
# 11. FINAL EVALUATION — TEST SET TOUCHED ONCE
# ============================================================
def evaluate_once(model, X_test_enc, y_test):
    y_pred = model.predict(X_test_enc)
    y_proba = model.predict_proba(X_test_enc)[:, 1] if len(model.classes_) == 2 else None

    report = classification_report(y_test, y_pred, output_dict=True)
    print(classification_report(y_test, y_pred))

    metrics = {"classification_report": report}

    if y_proba is not None:
        roc_auc = roc_auc_score(y_test, y_proba)
        precision, recall, _ = precision_recall_curve(y_test, y_proba, pos_label=model.classes_[1])
        pr_auc = auc(recall, precision)
        print(f"ROC-AUC: {roc_auc:.4f} | PR-AUC: {pr_auc:.4f}")
        metrics["roc_auc"] = roc_auc
        metrics["pr_auc"] = pr_auc

    return metrics


# ============================================================
# 12. FREEZE ARTIFACTS
# ============================================================
def save_artifacts(ensemble, preprocessor, feature_columns, metrics, dataset_hash):
    joblib.dump(ensemble, os.path.join(OUTPUT_DIR, "ensemble_model.pkl"))
    joblib.dump(preprocessor, os.path.join(OUTPUT_DIR, "preprocessor.pkl"))

    with open(os.path.join(OUTPUT_DIR, "feature_schema.json"), "w") as f:
        json.dump({"features": feature_columns}, f, indent=2)

    with open(os.path.join(OUTPUT_DIR, "model_metadata.json"), "w") as f:
        json.dump({
            "model_name": "ForenXAI_Ensemble_v1",
            "model_type": "SoftVotingClassifier",
            "training_dataset": "UWF-ZeekData24",
            "dataset_sha256": dataset_hash,
            "metrics": metrics,
            "random_state": RANDOM_STATE,
        }, f, indent=2, default=str)

    print(f"[+] Artifacts saved to {OUTPUT_DIR}/")


# ============================================================
# MAIN — run in order, in ONE cell to avoid execution-order bugs
# ============================================================
if __name__ == "__main__":
    dataset_hash = acquire_dataset()  # from Section 1, already established

    df = pd.read_csv(RAW_CSV_PATH, low_memory=False)
    profile_dataset(df)

    df = stratified_subsample(df)
    df = clean_data(df)
    X, y = split_features_labels(df)

    X_train, X_test, y_train, y_test = split_data(X, y)

    preprocessor = build_and_fit_preprocessor(X_train)
    X_train_enc = preprocessor.transform(X_train)
    X_test_enc = preprocessor.transform(X_test)

    X_train_res, y_train_res = apply_smote(X_train_enc, y_train)

    candidates, cv_results = train_and_cv_candidates(X_train_res, y_train_res)
    top_candidates = select_top_candidates(candidates, cv_results, top_n=3)

    ensemble = build_ensemble(top_candidates, X_train_res, y_train_res)
    metrics = evaluate_once(ensemble, X_test_enc, y_test)

    save_artifacts(ensemble, preprocessor, list(X.columns), metrics, dataset_hash)
```
